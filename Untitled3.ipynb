**Гипотеза 3**

**Нулевая гипотеза** Средняя стоимость на двоих в самых популярных ресторанах одинаковая H0\
**Альтернативная гипотеза** Средняя стоимость на двоих в самых популярных ресторанах разная H1

```python
result = st.ttest_ind(df.query('rating == 5')['cost for two'].dropna(),
                      df.query('rating == 5')['cost for two'].dropna())
prob_first1 = result.pvalue
print('p-значения', prob_first1)
alpha = .05
if prob_first1 < alpha:
    print('Отвергаем нулевую гипотеза')
else:
    print('Не получилось отвергнуть нулевую гипотеза')
```

Вывод: нулевая гипотеза подтвердилась, стоимость на двоих равна в ресторанах с рейтингом 5
